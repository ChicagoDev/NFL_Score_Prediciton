```python
def split_and_join_rush_stats(df):
    
    df['RUSH_ATT_challenger'] = df['Rush-Yds-TDs_challenger'].apply(lambda stat: int(stat.split('-')[0]))

    df['RUSH_YDS_challenger'] = df['Rush-Yds-TDs_challenger'].apply(lambda stat: int(stat.split('-')[1]))

    df['RUSH_TDs_challenger'] = df['Rush-Yds-TDs_challenger'].apply(lambda stat: int(stat.split('-')[2]))
    
    
    df['RUSH_ATT_defender'] = df['Rush-Yds-TDs_defender'].apply(lambda stat: int(stat.split('-')[0]))

    df['RUSH_YDS_defender'] = df['Rush-Yds-TDs_defender'].apply(lambda stat: int(stat.split('-')[1]))

    df['RUSH_TDs_defender'] = df['Rush-Yds-TDs_defender'].apply(lambda stat: int(stat.split('-')[2]))
    

    
def split_and_join_pass_stats(df):
    
    df['PASS_CMP_challenger'] = df['Cmp-Att-Yd-TD-INT_challenger'].apply(lambda stat: int(stat.split('-')[0]))

    df['PASS_ATT_challenger'] = df['Cmp-Att-Yd-TD-INT_challenger'].apply(lambda stat: int(stat.split('-')[1]))

    df['PASS_YDs_challenger'] = df['Cmp-Att-Yd-TD-INT_challenger'].apply(lambda stat: int(stat.split('-')[2]))
    
    df['PASS_TDs_challenger'] = df['Cmp-Att-Yd-TD-INT_challenger'].apply(lambda stat: int(stat.split('-')[3]))
    
    df['PASS_INT_challenger'] = df['Cmp-Att-Yd-TD-INT_challenger'].apply(lambda stat: int(stat.split('-')[4]))
    
    
    df['PASS_CMP_defender'] = df['Cmp-Att-Yd-TD-INT_defender'].apply(lambda stat: int(stat.split('-')[0]))

    df['PASS_ATT_defender'] = df['Cmp-Att-Yd-TD-INT_defender'].apply(lambda stat: int(stat.split('-')[1]))

    df['PASS_YDs_defender'] = df['Cmp-Att-Yd-TD-INT_defender'].apply(lambda stat: int(stat.split('-')[2]))
    
    df['PASS_TDs_defender'] = df['Cmp-Att-Yd-TD-INT_defender'].apply(lambda stat: int(stat.split('-')[3]))
    
    df['PASS_INT_defender'] = df['Cmp-Att-Yd-TD-INT_defender'].apply(lambda stat: int(stat.split('-')[4]))
    
    
```

## Splitting More Stats

```python
split_stats_df.iloc[0:5,21:30]
```

```python
# Sacked-Yards_challenger   2-13
# Fumbles-Lost_challenger    2-1
# Penalties-Yards_challenger  11-101

# Third Down Conv._challenger  8-16 -- Division
# Fourth Down Conv._challenger 0-1, 0/0 -- NaN Ratio ...  Just do Conversions...?
# 

def split_def_and_misc_stats(df):
    
    df['SACKS_challenger'] = df['Sacked-Yards_challenger'].apply(lambda stat: int(stat.split('-')[0]))
    df['SACK_YDs_challenger'] = df['Sacked-Yards_challenger'].apply(lambda stat: int(stat.split('-')[1]))

    df['SACKS_defender'] = df['Sacked-Yards_defender'].apply(lambda stat: int(stat.split('-')[0]))
    df['SACK_YDs_defender'] = df['Sacked-Yards_defender'].apply(lambda stat: int(stat.split('-')[1]))


    df['FUMBLES_challenger'] = df['Fumbles-Lost_challenger'].apply(lambda stat: int(stat.split('-')[0]))
    df['FUMBLES_LOST_challenger'] = df['Fumbles-Lost_challenger'].apply(lambda stat: int(stat.split('-')[1]))

    df['FUMBLES_defender'] = df['Fumbles-Lost_defender'].apply(lambda stat: int(stat.split('-')[0]))
    df['FUMBLES_LOST_defender'] = df['Fumbles-Lost_defender'].apply(lambda stat: int(stat.split('-')[1]))


    df['PENALTIES_challenger'] = df['Penalties-Yards_challenger'].apply(lambda stat: int(stat.split('-')[0]))
    df['PENALTY_YDs_challenger'] = df['Penalties-Yards_challenger'].apply(lambda stat: int(stat.split('-')[1]))

    df['PENALTIES_defender'] = df['Penalties-Yards_defender'].apply(lambda stat: int(stat.split('-')[0]))
    df['PENALTY_YDs_defender'] = df['Penalties-Yards_defender'].apply(lambda stat: int(stat.split('-')[1]))


    df['3rd_DWN_CONV_PCT_challenger'] = df['Third Down Conv._challenger'].apply(lambda stat: int(stat.split('-')[0]) / int(stat.split('-')[1]))
    df['3rd_DWN_CONV_PCT_defender'] = df['Third Down Conv._defender'].apply(lambda stat: int(stat.split('-')[0]) / int(stat.split('-')[1]))


```

```python
import copy

def split_the_extra_stats(df):
    
    splitting_db = copy.deepcopy(df)
    
    split_def_and_misc_stats(splitting_db)
    
    split_df = splitting_db.drop(['Sacked-Yards_challenger', \
                                  'Sacked-Yards_defender', \
                                  'Fumbles-Lost_challenger', \
                                  'Fumbles-Lost_defender',\
                                 'Penalties-Yards_challenger',\
                                 'Penalties-Yards_defender',
                                 'Third Down Conv._challenger'], axis=1)
        
    return split_df
```
